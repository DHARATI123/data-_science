```python
dataframe = pd.DataFrame(bos1) 
  
print("Given Dataframe :\n", dataframe)  
  
# selecting rows based on condition 
rslt_df = dataframe.loc[dataframe['No. of Inmates benefitted by Elementary Education'] != 0] 
```

```python
col_list= list(dataframe)
col_list.remove('STATE/UT')
col_list.remove('YEAR')
col_list
```

```python
dataframe['total_benefitted'] = dataframe[col_list].sum(axis=1)
dataframe
```

```python
dataframe.loc['totals']= dataframe.sum(numeric_only=True, axis=0)
dataframe
```

```python
dataframe.drop(dataframe.tail(1).index,inplace=True)
fig = plt.figure()
ax = fig.add_axes([0,0,2,2])
state = dataframe['STATE/UT'] 
total_inmates = dataframe['total_benefitted'] 
ax.bar(state,total_inmates)
plt.show()
```

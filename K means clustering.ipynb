```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (16,9)
plt.style.use('ggplot')
```

```python
df = pd.read_csv('movie_metadata.csv')
df.head()
```

```python
df['director_facebook_likes']
```

```python
df.columns
```

```python
X = df.iloc[:,4:6]
X
```

```python
zero_not_accepted = ['director_facebook_likes','actor_3_facebook_likes']

for column in zero_not_accepted:
    X[column] = X[column].replace(0, np.NaN)
    mean = int(X[column].mean(skipna=True))
    X[column] = X[column].replace(np.NaN, mean)
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters = 5)
```

```python
kmeans.fit(X)
```

```python
kmeans.cluster_centers_
```

```python
print(kmeans.labels_)
print(len(kmeans.labels_))
```

```python
print(type(kmeans.labels_))
unique , counts = np.unique(kmeans.labels_ , return_counts=True )
print(dict(zip(unique , counts)))
```

```python
X['cluster'] = kmeans.labels_
sns.set_style('whitegrid')
sns.lmplot('director_facebook_likes','actor_3_facebook_likes' , data =X , hue = 'cluster' ,
           palette = 'coolwarm',size =6,aspect =1, fit_reg = False)
```
